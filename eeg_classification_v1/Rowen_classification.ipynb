```python
import scipy.io
import numpy as np
import sklearn
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data= np.loadtxt(r"C:\Users\rowen\Documents\GitHub\SN_2023\eeg_classification_v1\all_fouriers", delimiter=",")
                  
print(np.shape(data[0]))
x = data[:,:-1]
y = data[:,-1]
print(np.shape(x))
print(np.shape(y))
print(y.reshape(1885,1))

```

```python
from sklearn import svm



trainfeatures = x[0:1500]
trainlabels = y[0:1500]

testfeatures = x[1500:]
testlabels = y[1500:]


rbf_svc = svm.SVC()


rbf_svc.fit(trainfeatures, trainlabels)

test = rbf_svc.predict(testfeatures)
count = 0
for i in range(385): 
    if testlabels[i] == test[i]:
        count += 1

accuracy = count / 385
print("accuracy is ", accuracy*100,"%")

```

```python
import matplotlib.pyplot as plt

# Plot histogram of true labels
plt.hist(testlabels, bins=10, alpha=0.5, label='True Labels')
# Plot histogram of predicted labels
plt.hist(test, bins=10, alpha=0.5, label='Predicted Labels')

plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Histogram of True vs Predicted Labels')
plt.legend()

plt.show()
```
