```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
data1 = np.zeros((1000,3))
data2 = np.zeros((1000,3))

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-1,high=1)
    y = np.random.uniform(low=-1,high=1)
    if (x**2 + y**2)<1:
        data1[counter,0] = x
        data1[counter,1] = y
        data1[counter,2] = 1
        counter+=1

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-2,high=2)
    y = np.random.uniform(low=-2,high=2)
    if (x**2 + y**2)>1:
        if (x**2 + y**2)<3:
            data2[counter,0] = x
            data2[counter,1] = y
            data2[counter,2] = 2
            counter+=1


practice1 = np.concatenate((data1,data2))
np.random.shuffle(practice1)

print(practice1)

plt.scatter(data1[:,0], data1[:,1],color="red", alpha=0.3)
plt.scatter(data2[:,0], data2[:,1],color="blue",alpha=0.3)
```

```python
from sklearn import svm

trainfeatures3 = practice1[0:1200, 0:2]
trainlabels3 = practice1[0:1200, 2]

testfeatures3 = practice1[1200:, 0:2]
testlabels3 = practice1[1200:, 2]


rbf_svc = svm.SVC(kernel='rbf')
rbf_svc.kernel


rbf_svc.fit(trainfeatures3, trainlabels3)

test = rbf_svc.predict(testfeatures3)
count = 0
for i in range(800): 
    if testlabels3[i] == test[i]:
        count += 1

accuracy = count / 800
print("accuracy is ", accuracy*100,"%")
```

```python
data1 = np.zeros((1000,4))
data2 = np.zeros((1000,4))
data3 = np.zeros((1000,4))


x1 = 3
y1 = -2
z1 = -1
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data1[counter,0] = x2
        data1[counter,1] = y2
        data1[counter,2] = z2
        data1[counter,3] = 1
        counter+=1

x1 = -2
y1 = -1
z1 = -3
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data2[counter,0] = x2
        data2[counter,1] = y2
        data2[counter,2] = z2
        data2[counter,3] = 2
        counter+=1

x1 = -1
y1 = 2
z1 = -1.5
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data3[counter,0] = x2
        data3[counter,1] = y2
        data3[counter,2] = z2
        data3[counter,3] = 3
        counter+=1

practice2 = np.concatenate((data1,data2,data3))
np.random.shuffle(practice2)
print(practice2)

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data1[:,0], data1[:,1], data1[:,2], color = "green",alpha=0.2)
ax.scatter3D(data2[:,0], data2[:,1], data2[:,2], color = "blue",alpha=0.2)
ax.scatter3D(data3[:,0], data3[:,1], data3[:,2], color = "red",alpha=0.2)
plt.show()
```

```python
from sklearn import svm

trainfeatures2 = practice2[0:2000, 0:3]
trainlabels2 = practice2[0:2000, 3]

testfeatures2 = practice2[2000:, 0:3]
testlabels2 = practice2[2000:, 3]

classifier = svm.SVC()

classifier.fit(trainfeatures2, trainlabels2)

test = classifier.predict(testfeatures2)
count = 0
for i in range(1000): 
    if testlabels2[i] == test[i]:
        count += 1

accuracy = count / 1000
print("accuracy is ", accuracy*100,"%")
```
