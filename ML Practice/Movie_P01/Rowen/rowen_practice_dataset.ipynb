```python
import scipy.io
import numpy as np

# Load EEG data
raw_data = {}
for i in range(1, 37):
    mat = scipy.io.loadmat("EEG_Clip" + str(i) + ".mat")
    raw_data[i] = mat["ThisEEG"]

# Load self-report ratings
emotions = scipy.io.loadmat("Dt_SelfReports.mat")
ratings_array = emotions["Ratings"]

# Transpose the array to have movies as the first dimension
ratings_array = np.transpose(ratings_array, axes=(2, 0, 1))

# Iterate through each emotion
for emotion_index in range(ratings_array.shape[1]):
    emotion_ratings = ratings_array[:, emotion_index, :]

    # Iterate through each movie
    for movie_index in range(emotion_ratings.shape[0]):
        movie_rating = emotion_ratings[movie_index, :]

        # Calculate average rating for the current emotion and movie
        average_rating = np.mean(movie_rating)

        # Print or store the result as needed
        print(f"Average Rating for Movie {movie_index + 1}, Emotion {emotion_index + 1}: {average_rating}")
```

```python
from sklearn import svm
import scipy.io
import numpy as np

raw_data = {}
for i in range(1, 37):
    mat = scipy.io.loadmat("EEG_Clip" + str(i) + ".mat")
    raw_data[i] = mat["ThisEEG"]

emotions = scipy.io.loadmat("Dt_SelfReports.mat")

trainfeatures = emotions[0:40, 0:1]
trainlabels = emotions[0:40, 1]

testfeatures = emotions[40:, 0:1]
testlabels = emotions[40:, 1]

classifier = svm.SVC()

classifier.fit(trainfeatures, trainlabels)

test = classifier.predict(testfeatures)
count = 0
for i in range(18): 
    if testlabels[i] == test[i]:
        count += 1

accuracy = count / 18
print("accuracy is ", accuracy*100,"%")
```

```python
from sklearn import svm
import scipy.io
import numpy as np

# Load EEG data
raw_data = {}
for i in range(1, 37):
    mat = scipy.io.loadmat("EEG_Clip" + str(i) + ".mat")
    raw_data[i] = mat["ThisEEG"]

# Load self-report ratings
emotions = scipy.io.loadmat("Dt_SelfReports.mat")

# Assuming the structure of your data is (movies, emotions, people)
ratings_array = emotions["Ratings"]

# Reshape the ratings_array for compatibility with SVM
ratings_array_reshaped = ratings_array.reshape((ratings_array.shape[0], -1))

# Assuming you have 58 movies
num_movies = ratings_array_reshaped.shape[0]
num_train_movies = 40

# Use the first 40 movies for training
trainfeatures = ratings_array_reshaped[:num_train_movies]
trainlabels = emotions["Ratings"][:, 1][:num_train_movies]

# Use the remaining movies for testing
testfeatures = ratings_array_reshaped[num_train_movies:]
testlabels = emotions["Ratings"][:, 1][num_train_movies:]

# Handle missing values (replace NaN with the mean of the column)
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
trainfeatures_imputed = imputer.fit_transform(trainfeatures)
testfeatures_imputed = imputer.transform(testfeatures)

# Create the SVM classifier
classifier = svm.SVC()

# Fit the classifier on the training data
classifier.fit(trainfeatures_imputed, trainlabels)

# Predict on the test data
test = classifier.predict(testfeatures_imputed)

# Calculate accuracy
accuracy = np.sum(testlabels == test) / len(testlabels)
print("Accuracy is {:.2%}".format(accuracy))
```
