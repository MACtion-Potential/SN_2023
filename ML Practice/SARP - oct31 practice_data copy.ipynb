```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn

from sklearn import svm
```

```python
data1 = np.zeros((1000,3))
data2 = np.zeros((1000,3))

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-1,high=1)
    y = np.random.uniform(low=-1,high=1)
    if (x**2 + y**2)<1:
        data1[counter,0] = x
        data1[counter,1] = y
        data1[counter,2] = 1
        counter+=1

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-2,high=2)
    y = np.random.uniform(low=-2,high=2)
    if (x**2 + y**2)>1:
        if (x**2 + y**2)<3:
            data2[counter,0] = x
            data2[counter,1] = y
            data2[counter,2] = 2
            counter+=1


practice1 = np.concatenate((data1,data2))
np.random.shuffle(practice1)

print(practice1)

plt.scatter(data1[:,0], data1[:,1],color="red", alpha=0.3)
plt.scatter(data2[:,0], data2[:,1],color="blue",alpha=0.3)
```

```python
data1 = np.zeros((1000,4))
data2 = np.zeros((1000,4))
data3 = np.zeros((1000,4))


x1 = 3
y1 = -2
z1 = -1
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data1[counter,0] = x2
        data1[counter,1] = y2
        data1[counter,2] = z2
        data1[counter,3] = 1
        counter+=1

x1 = -2
y1 = -1
z1 = -3
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data2[counter,0] = x2
        data2[counter,1] = y2
        data2[counter,2] = z2
        data2[counter,3] = 2
        counter+=1

x1 = -1
y1 = 2
z1 = -1.5
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data3[counter,0] = x2
        data3[counter,1] = y2
        data3[counter,2] = z2
        data3[counter,3] = 3
        counter+=1

practice2 = np.concatenate((data1,data2,data3))
np.random.shuffle(practice2)
print(practice2)

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data1[:,0], data1[:,1], data1[:,2], color = "green",alpha=0.2)
ax.scatter3D(data2[:,0], data2[:,1], data2[:,2], color = "blue",alpha=0.2)
ax.scatter3D(data3[:,0], data3[:,1], data3[:,2], color = "red",alpha=0.2)
plt.show()
```

```python
from sklearn.svm import SVC

# Example data (X) and labels (y)
X = [[1.0, 2.0], [2.0, 3.0], [3.0, 4.0]]
y = [0, 1, 0]  # labels as integers (0 and 1)

# Create an SVM classifier
clf = svm.SVC();

# Train the classifier on your data
clf.fit(X, y)

X_new = [[4.0, 5.0], [1.0, 2.0]]
predicted_labels = clf.predict(X_new)
print("Predicted labels:", predicted_labels)
```

```python
from sklearn.svm import SVC

# First dataset for training
data1 = [
    [-0.55681372, -0.14521181, 1.0],
    [-0.51086784, 1.33389013, 2.0],
    [-0.33517242, 0.97016578, 2.0],
]

# Second dataset for evaluation (leaving the last set out)
data2 = [
    [-1.89855022, 1.35549277, -1.19678493, 3.0],
    [-2.84957672, -1.41839909, 1.0],
    [-2.27715208, -0.48849317, 1.0],
]

X_train = [row[:2] for row in data1]  # Features (x, y)
y_train = [row[2] for row in data1]   # Labels

# Create an SVM classifier
clf = SVC()

# Train the classifier on the first dataset
clf.fit(X_train, y_train)

# Make predictions on the second dataset
predicted_labels = clf.predict(data2)

# Print the predicted labels
print("Predicted labels for the second dataset:", predicted_labels)
```
