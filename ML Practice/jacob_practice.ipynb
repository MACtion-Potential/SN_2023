```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm,datasets
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
```

# Defining datasets

```python
data1 = np.zeros((1000,3))
data2 = np.zeros((1000,3))

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-1,high=1)
    y = np.random.uniform(low=-1,high=1)
    if (x**2 + y**2)<1:
        data1[counter,0] = x
        data1[counter,1] = y
        data1[counter,2] = 1
        counter+=1

counter = 0
while counter < 1000:
    x = np.random.uniform(low=-2,high=2)
    y = np.random.uniform(low=-2,high=2)
    if (x**2 + y**2)>1:
        if (x**2 + y**2)<3:
            data2[counter,0] = x
            data2[counter,1] = y
            data2[counter,2] = 2
            counter+=1


practice1 = np.concatenate((data1,data2))
np.random.shuffle(practice1)

print(practice1)

plt.scatter(data1[:,0], data1[:,1],color="red", alpha=0.3)
plt.scatter(data2[:,0], data2[:,1],color="blue",alpha=0.3)
```

```python
data1 = np.zeros((1000,4))
data2 = np.zeros((1000,4))
data3 = np.zeros((1000,4))


x1 = -2
y1 = -1
z1 = -3
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data1[counter,0] = x2
        data1[counter,1] = y2
        data1[counter,2] = z2
        data1[counter,3] = 1
        counter+=1

x1 = 3
y1 = -2
z1 = -1
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data2[counter,0] = x2
        data2[counter,1] = y2
        data2[counter,2] = z2
        data2[counter,3] = 2
        counter+=1

x1 = -1
y1 = 2
z1 = -1.5
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<2:
        data3[counter,0] = x2
        data3[counter,1] = y2
        data3[counter,2] = z2
        data3[counter,3] = 3
        counter+=1

practice2 = np.concatenate((data1,data2,data3))
np.random.shuffle(practice2)
print(practice2)

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data1[:,0], data1[:,1], data1[:,2], color = "green",alpha=0.2)
ax.scatter3D(data2[:,0], data2[:,1], data2[:,2], color = "blue",alpha=0.2)
ax.scatter3D(data3[:,0], data3[:,1], data3[:,2], color = "red",alpha=0.2)
plt.show()
```

```python
data1 = np.zeros((1000,4))
data2 = np.zeros((1000,4))

x1 = 3
y1 = -2
z1 = -1
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<1:
        data1[counter,0] = x2
        data1[counter,1] = y2
        data1[counter,2] = z2
        data1[counter,3] = 1
        counter+=1

x1 = 3
y1 = -2
z1 = 0
counter = 0
while counter < 1000:
    x2 = np.random.uniform(low=-1,high=1) + x1
    y2= np.random.uniform(low=-1,high=1) + y1
    z2 = np.random.uniform(low=-1,high=1) + z1 
    if ((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)<1:
        data2[counter,0] = x2
        data2[counter,1] = y2
        data2[counter,2] = z2
        data2[counter,3] = 2
        counter+=1


practice3 = np.concatenate((data1,data2))
np.random.shuffle(practice3)
print(practice3)

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data1[:,0], data1[:,1], data1[:,2], color = "green",alpha=0.2)
ax.scatter3D(data2[:,0], data2[:,1], data2[:,2], color = "blue",alpha=0.2)
plt.show()
```

# Support Vector Machines

```python
#training a support vector machine to classify practice3

#taking the x,y coordinates from the first 1800 points
training_features = practice3[0:1800,0:3]
#taking the labels from the first 1800 points
training_labels = practice3[0:1800,3]

#taking coords from 200 last points
testing_features = practice3[1800:,0:3]
#taking labels from 200 last points
testing_labels = practice3[1800:,3]

machine  = svm.SVC()

machine.fit(training_features, training_labels)
```

```python
testing = machine.predict(testing_features)
count = 0
for i in range(200):
    if testing_labels[i] == testing[i]:
        count += 1

print("accuracy=", count/200)
```

# Neural Networks

```python
#taking the x,y coordinates from the first 1800 points
training_features = practice2[0:2500,0:3]
#taking the labels from the first 1800 points
training_labels = practice2[0:2500,3]

#taking coords from 200 last points
testing_features = practice2[2500:,0:3]
#taking labels from 200 last points
testing_labels = practice2[2500:,3]

#defining the neural network
#"solver" is the algorithm that the network uses to train its hidden weights... not sure on the differences b/w them tbh
#"alpha" is the "strength of the L2 regularization term", not really sure what this does, I think it removes a little bit from each weight each iteration, can help with overfitting maybe it says?
#"random state" sets the seed that the weights and biases are randomly drawn from, if you set it you will get the same results each time you run the code, otherwise there may be some variation
#"hidden layers" is a vector where the ith entry describes the number of units in the ith hidden layer, thus hidden_layer_sies = (5,2) would give 2 hidden layers of 5 and 2 units respectivly
n_network = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(2), random_state=1)

n_network.fit(training_features,training_labels)

testing = n_network.predict(testing_features)
count = 0
for i in range(500):
    if testing_labels[i] == testing[i]:
        count += 1

print("accuracy=", count/500)
```

```python
#attempting to do some PCA analysis
#this is data from the sklearn website, we are just importing and plotting it

iris = datasets.load_iris()

#print(iris.data)
#print(iris.data[:,0])
#print(iris.data[:,1])

#stealing this from the sklearn website, each flower has 5 characteristics, here we are plotting two of them, you can see there is some variation
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
#this is the actual PCA analysis, we are collapsing the variability along the 5 dimensions into just 3 dimensions, however, now the actual dimensions are meaningless, but we can see there is variability within the flowers

import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=iris.target,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])
```
